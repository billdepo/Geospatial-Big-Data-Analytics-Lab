```python
import numpy as np
import matplotlib.pyplot as plt
from sklearn.metrics import confusion_matrix, ConfusionMatrixDisplay

def plot_conf_matrix(classes = [0, 1, 2, 3], cm = [[9,0,0,0],[0,2,0,0],[1,0,6,0],[0,1,0,2]]):
    conf_matrix = np.array(cm)
    disp = ConfusionMatrixDisplay(confusion_matrix=conf_matrix, display_labels=classes)
    disp.plot()
    plt.show()
```

```python
plot_conf_matrix(classes = [0, 1, 2], cm = [[2,0,0],[0,5,1],[0,1,5]]) # CART Confusion matrix
```

```python
plot_conf_matrix(classes = [0, 1, 2], cm = [[2,0,0],[0,6,0],[0,0,6]]) # SVM Confusion matrix
```
